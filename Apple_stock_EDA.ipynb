```python
from pandas.plotting import lag_plot
# create a scatter plot
for i in [1, 4, 8, 12, 16, 20, 24, 38, 32, 36, 40]:
    #plt.figure(figsize=(12,8))
    lag_plot(df.Close, lag=i)
    plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose= seasonal_decompose(df.Close,period=21)
decompose.plot()
plt.rcParams['figure.figsize'] = (30, 15);
```

```python
decompose= seasonal_decompose(df.Close,period=63)
decompose.plot()
plt.rcParams['figure.figsize'] = (30, 15);
```

```python
decompose= seasonal_decompose(df.Close,period=252)
decompose.plot()
plt.rcParams['figure.figsize'] = (30, 15);
```

Data is having seasonality and exponential growth in Trend, so we can say that data is non stationary
residuals doesn't have much deviation and there is no outliers

```python
df["Close"].rolling(1, center=False).mean().plot(legend=True, label="Rolling Mean");
print("Mean is:", df["Close"].mean())
```

```python
from statsmodels.tsa.stattools import adfuller
adf = adfuller(df["Close"])[1]
print(f"p value:{adf.round(4)}", ", Series is Stationary" if adf <0.05 else ", Series is Non-Stationary")
```

after seeing lag plots, rolling mean and adfuller test
we can say that data is not stationary

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df.Close)
plot_pacf(df.Close)
plt.show() 
```

```python
plot_acf(df.Close.diff(21).dropna())
plot_pacf(df.Close.diff(21).dropna())
plt.show() 
```

```python
adfuller(df.Close.diff(1).dropna())
```

```python
plot_acf(df.Close.diff(1).dropna())
plot_pacf(df.Close.diff(1).dropna())
plt.show() 
```

```python
plt.rcParams.update({'figure.figsize':(9,9), 'figure.dpi':120})

fig, axes = plt.subplots(3, 2)
axes[0, 0].plot(df.Close); axes[0, 0].set_title('Original Series')
plot_acf(df.Close, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(df.Close.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df.Close.diff().dropna(), ax=axes[1, 1])

# seasonal Differencing
axes[2, 0].plot(df.Close.diff(21)); axes[2, 0].set_title('seasonal Differencing')
plot_acf(df.Close.diff(21).dropna(), ax=axes[2, 1])

plt.show()
```
